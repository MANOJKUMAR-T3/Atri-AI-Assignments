```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
(train_images, train_labels), _ = fashion_mnist.load_data()
```

```python
fa_class = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]
chos_imag = [None] * 10
for i, j in zip(train_images, train_labels):
    if chos_imag[j] is None:
        chos_imag[j] = i
    if all(img is not None for img in chos_imag):
        break
```

```python
fig, a = plt.subplots(2, 5, figsize=(12, 5))
for i, j in enumerate(a.flat):
    j.imshow(chos_imag[i], cmap="gray")
    j.set_title(fa_class[i])
    j.set_xticks([])
    j.set_yticks([])
fig.suptitle("Fashion-MNIST: Sample Image per Class", fontsize=13)
plt.tight_layout()
plt.show()
```
